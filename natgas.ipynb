Each point in the data set corresponds to the purchase price of natural gas at the end of a month, from 31st October 2020 to 30th September 2024.
Analyze the data to estimate the purchase price of gas at any date in the past and extrapolate it for one year into the future. 
Your code should take a date as input and return a price estimate.


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
natural_gas = pd.read_csv("Nat_Gas.csv")

#see how the data looks like, first 5 rows

natural_gas.head(10)
```

```python
natural_gas.info()
```

```python
# Line plot to visualise how it looks like
natural_gas.plot("Dates", "Prices")
```

```python
import datetime
#import this date time to convert to datetime format
#lets properly convert the column to datetime
natural_gas["Dates"] = pd.to_datetime(natural_gas["Dates"])
```

```python
natural_gas.info()
```

```python
#lets try to plot it using seaborn

sns.set(style='whitegrid')

plt.figure(figsize=(12,6))
sns.lineplot(data=natural_gas, x = natural_gas["Dates"], y = natural_gas["Prices"],
             label = "Natural Gas Prices Monthly", color="green")

plt.xlabel("Dates (Monthly)")
plt.ylabel("Prices (usd?)")

plt.show()
```
